# Python Functions : Using `def`

## What is a Function?
A **function** in Python is a reusable block of code that performs a specific task.  
It helps break down complex problems into smaller parts, improves readability, and avoids repetition.

Functions are defined using the **`def` keyword**.

---

## Syntax of a Function
```python
def function_name(parameters):
    """docstring (optional description)"""
    # function body
    return result


- def → Keyword to define a function.
- function_name → Identifier used to call the function.
- parameters → Inputs passed into the function.
- return → Sends output back to the caller.

````python
def greet():
    return "Hello, World!"
````

````python
print(greet())  
````

````python
# Example 2: Function with One Parameter
def square(x):
    return x * x
````

````python
print(square(5))  
````

````python
# Example 3: Function with Two Parameters
def add(a, b):
    return a + b
````

````python
print(add(3, 7))  
````

````python
# Example 4: Function with Conditional Logic
def is_even(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"
````

````python
print(is_even(4)) 
````

````python
# Example 5: Function Returning Multiple Values
def calc(a, b):
    return a + b, a * b
````

````python
print(calc(2, 3)) 
````

#  Understanding `*args` in Python

## What are `*args`?
- In Python, `*args` allows a function to accept **any number of positional arguments**.
- The arguments are collected into a **tuple** inside the function.
- Useful when you don’t know in advance how many inputs will be passed.

---

##  Syntax
```python
def function_name(*args):
    # function body


````python
# Example 1: Printing All Arguments

def show_items(*args):
    print(args)
````

````python
show_items(10, 20, 30)
````

````python
# Example 2: Looping Through Arguments
def print_items(*args):
    for item in args:
        print(item)
````

````python
print_items("apple", "banana", "cherry")
````

````python
# Example 3: Summing Numbers

def add_numbers(*args):
    return sum(args)
````

````python
print(add_numbers(5, 10, 15))
````

# Lambda Functions in Python

## 1. What is a Lambda Function?
- A **lambda function** is a small, anonymous function in Python.
- It is defined using the keyword `lambda` instead of `def`.
- Unlike normal functions, a lambda function has **no name** and is usually used for short, simple operations.

**Syntax:**


- `arguments`: Input values (like parameters in a normal function).
- `expression`: A single statement that is evaluated and returned.


lambda arguments: expression

---

## 2. Key Characteristics
- **Anonymous**: No function name is required.
- **Single Expression**: Can only contain one expression (no multiple statements).
- **Inline Use**: Often used where functions are needed temporarily.

---

## 3. Examples

### Example 1: Simple Addition
```python
add = lambda x, y: x + y
print(add(5, 3))  # Output: 8

## 4. When to Use Lambda?

When you need a short function for a quick task.

Commonly used with functions like map(), filter(), and sorted().

Helps keep code concise and readable.

## 5. Limitations

Cannot contain multiple expressions or statements.

Not suitable for complex logic, use def for larger functions.

### Examples:

````python
# 1 Lambda with no parameters (constant return)
hello = lambda: "Hello, World!"
````

````python
print(hello()) 
````

````python
# 2 Lambda with one parameter (square of a number)
square = lambda x: x * x
````

````python
print(square(5))  
````

````python
# 3 Lambda with two parameters (addition)
add = lambda a, b: a + b
````

````python
print(add(3, 7))  
````

````python
def check_odd(x):
    if x % 2 == 0:
        print("Even")
    else:
        print("odd")

check_odd(1)
````
